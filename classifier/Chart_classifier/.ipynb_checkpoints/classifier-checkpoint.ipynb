```python
def resize_images(path):
    directory = path

    for file_name in os.listdir(directory):
      print("Processing %s" % file_name)
      image = Image.open(os.path.join(directory, file_name))

      x,y = image.size
      new_dimensions = (200, 200)
      output = image.resize(new_dimensions, Image.ANTIALIAS)

      output_file_name = os.path.join(directory, file_name)
      output.save(output_file_name, "PNG", quality = 100)

    print("All done")
```

```python
def load_data(csv_path):
    image_data = []
    label_data = []
    df = pd.read_csv(csv_path, sep=",", index_col=False)
    data_len = len(df['Image_path'])
    for i in range(data_len):
        temp = cv2.imread(df['Image_path'][i])
        temp = (cv2.cvtColor(temp, cv2.COLOR_BGR2RGB)).tolist()
        image_data.append(temp)
        if df['Class'][i] == 'bar':
            label_data.append([1,0])
        else:
            label_data.append([0,1])
    return (image_data, label_data)
        
```

```python

X_test, y_test = load_data("/home/komaldadhich/Documents/study/Research/graph_percept/source/project/chart_percept/Chart_classifier/dataset/test/labels.csv")
```

```python
X_train, y_train = load_data("/home/komaldadhich/Documents/study/Research/graph_percept/source/project/chart_percept/Chart_classifier/dataset/train/labels.csv")
```

```python
X_train = np.array(X_train)
X_test = np.array(X_test)
y_train = np.array(X_train)
y_test = np.array(X_test)
print(X_test.shape)
print(X_train.shape)
```

```python
X_train = X_train.reshape(451,200,200,3)
X_test = X_test.reshape(149,200,200,3)
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
X_train /= 255.0
X_test /= 255.0
# y_train = to_categorical(y_train,num_classes)
# y_test = to_categorical(y_test, num_classes)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
cnn = Sequential()
cnn.add(Conv2D(16, kernel_size=(5,5),input_shape=(200,200,3), padding='same', activation='relu', data_format='channels_last'))
# cnn.add(Activation('relu'))
cnn.add(MaxPooling2D())
cnn.add(Conv2D(32, kernel_size=(5,5),padding='same', activation='relu'))
# cnn.add(Activation('relu'))
cnn.add(MaxPooling2D())
cnn.add(Flatten())        
cnn.add(Dense(4, activation='relu'))
# cnn.add(Flatten()) 
# cnn.add(Activation('relu'))
# cnn.add(Dropout(0.25))
cnn.add(Dense(2, activation='softmax'))
cnn.compile(loss='binary_crossentropy', optimizer='rmsprop', metrics=['accuracy'])
cnn.summary()
```

```python
history_cnn = cnn.fit(X_train,y_train,epochs=5,verbose=1,validation_data=(X_train,y_train))
```
